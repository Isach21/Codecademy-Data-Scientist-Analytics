# Exploring Mushrooms

## Looking Over the Data

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

```

```python
# load in the data
df = pd.read_csv("mushroom_data.csv")
print(df.head())
```

```python
# list of all column headers
columns = df.columns.tolist()
print(columns)
print(df.info())
```

```python
# graphs for each column
for i in columns: 
    sns.countplot(x=i, data=df)
    plt.show()
    plt.clf()

```

## Cleaning the Bar Graphs

```python
for i in columns: 
    sns.countplot(x=i, data=df)
    # rotates the value labels slightly so they don't overlap, also slightly increases font size
    plt.xticks(rotation=30, fontsize=10)
    # increases the variable label font size slightly to increase readability
    plt.xlabel(i, fontsize=12)    
    plt.title( i +  " Value Counts")
    plt.show()
    plt.clf()
```

## Ordering the Bars for Analysis

```python
for i in columns: 
    sns.countplot(x=i, data=df, order=df[i].value_counts().index)
    # rotates the value labels slightly so they don't overlap, also slightly increases font size
    plt.xticks(rotation=30, fontsize=10)
    # increases the variable label font size slightly to increase readability
    plt.xlabel(i, fontsize=12)    
    plt.title( i +  " Value Counts")
    plt.show()
    plt.clf()
```
